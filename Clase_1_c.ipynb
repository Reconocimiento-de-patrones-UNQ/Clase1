# Introducción a la visualización de datos con Pandas, Matplotlib y Seaborn

```python
# Cargar bibliotecas
import pandas as pd

# Dataset Iris incluido en seaborn, pero lo cargamos desde seaborn para usarlo con pandas
import seaborn as sns

# Cargamos el dataset Iris
iris = sns.load_dataset('iris')

# Mostrar las primeras filas
print (iris.sample(10))
```

```python
# Usamos el método .plot() de Pandas para un gráfico rápido

iris.plot(kind='scatter', x='sepal_length', y='sepal_width', title='Gráfico con pandas')
```

Nota: pandas nos permite crear gráficos simples de forma rápida. Es muy útil para una exploración inicial de los datos, aunque tiene menos opciones de personalización que matplotlib o seaborn.

##Graficar con matplotlib

```python
import matplotlib.pyplot as plt

# Creamos un gráfico similar pero con más control
plt.figure(figsize=(8, 5))
plt.scatter(iris['sepal_length'], iris['sepal_width'], c='purple', alpha=0.7)
plt.title('Gráfico con matplotlib')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, label='sin(x)')
plt.title('Gráfico de línea')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
values = [23, 45, 56, 78]

plt.bar(categories, values, color='skyblue')
plt.title('Gráfico de barras')
plt.ylabel('Valores')
plt.show()
```

```python
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Gráfico de torta')
plt.axis('equal')  # Para que sea un círculo
plt.show()
```

```python
import seaborn as sns
# Usamos seaborn para graficar con estilo más moderno y opciones estadísticas integradas
sns.set(style='whitegrid')  # Estilo más limpio
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species', palette='Set2')
plt.title('Gráfico con seaborn')
plt.show()
```

```python
import numpy as np
import pandas as pd

np.random.seed(42)
x = np.linspace(0, 10, 100)
y = 3*x + 5 + np.random.normal(0, 2, size=100)

df = pd.DataFrame({'x': x, 'y': y})
```

```python
from sklearn.linear_model import LinearRegression

X = df[['x']]  # sklearn espera una matriz 2D
y = df['y']

model_skl = LinearRegression()
model_skl.fit(X, y)

print(f"Coeficiente: {model_skl.coef_[0]:.3f}")
print(f"Intercepto: {model_skl.intercept_:.2f}")
```

Ventajas scikit-learn:

    Enfocado en machine learning.

    Muy rápido y optimizado.

    Facilita validación cruzada, pipelines y predicciones.

```python
import statsmodels.api as sm

X_sm = sm.add_constant(X)  # Necesita agregar la constante

model_sm = sm.OLS(y, X_sm).fit()

print(model_sm.summary())
```

Ventajas statsmodels:

    Reporta métricas estadísticas: p-valores, IC, R² ajustado.

    Ideal para ciencia de datos, análisis exploratorios y publicaciones.

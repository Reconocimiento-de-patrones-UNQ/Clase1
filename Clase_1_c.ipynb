##Graficar con matplotlib

```python
import matplotlib.pyplot as plt

# Creamos un gráfico similar pero con más control
plt.figure(figsize=(8, 5))
plt.scatter(iris['sepal_length'], iris['sepal_width'], c='purple', alpha=0.4)
plt.title('Gráfico con matplotlib')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, label='sin(x)')
plt.title('Gráfico de línea')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
values = [23, 45, 56, 78]

plt.bar(categories, values, color='skyblue')
plt.title('Gráfico de barras')
plt.ylabel('Valores')
plt.show()
```

```python
import seaborn as sns
# Usamos seaborn para graficar con estilo más moderno y opciones estadísticas integradas
sns.set(style='whitegrid')  # Estilo más limpio
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species', palette='Set2', alpha=0.9)
plt.title('Gráfico con seaborn')
plt.show()
```

##Tanto Scikit-learn como Statsmodels
permiten ajustar modelos de regresión lineal, pero difieren en enfoque y funcionalidades. Scikit-learn está orientado a machine learning, priorizando velocidad, simplicidad y uso en pipelines predictivos; su clase LinearRegression devuelve coeficientes y predicciones, pero no ofrece análisis estadístico detallado. En cambio, Statsmodels sigue una tradición más estadística: su función OLS proporciona resultados completos, incluyendo errores estándar, intervalos de confianza, valores p y medidas de ajuste como R² ajustado y AIC, lo que la hace ideal para análisis exploratorio e inferencial. En resumen, Scikit-learn es más adecuada para modelos predictivos, mientras que Statsmodels es preferida para interpretación y análisis estadístico.

```python
import numpy as np
import pandas as pd

np.random.seed(42)
x = np.linspace(0, 10, 100)
y = 3*x + 5 + np.random.normal(0, 2, size=100)

df = pd.DataFrame({'x': x, 'y': y})
df
```

```python
from matplotlib import pyplot as plt
df.plot(kind='scatter', x='x', y='y', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
from sklearn.linear_model import LinearRegression

X = df[['x']] # matriz 2D por algún motivo
#print(X)
y = df['y']

model_skl = LinearRegression()
model_skl.fit(X, y)

print(f"Coeficiente: {model_skl.coef_[0]:.3f}")
print(f"Intercepto: {model_skl.intercept_:.2f}")
```

Ventajas scikit-learn:

    Enfocado en machine learning.

    Muy rápido y optimizado.

    Facilita validación cruzada, pipelines y predicciones.

```python
import statsmodels.api as sm

X_sm = sm.add_constant(X)  # Necesita agregar la constante
#print (X_sm)
model_sm = sm.OLS(y, X_sm).fit()

print(model_sm.summary())
```

Ventajas statsmodels:

    Reporta métricas estadísticas: p-valores, IC, R² ajustado.

    Ideal para ciencia de datos, análisis exploratorios y publicaciones.

```python
plt.figure(figsize=(12, 5))

# Gráfico del ajuste
plt.subplot(1, 2, 1)
plt.scatter(X.iloc[:, 0], y, alpha=0.7, label='Datos')
plt.plot(X.iloc[:, 0], model_sm.fittedvalues, color='red', label='Ajuste')
plt.xlabel(X.columns[0])
plt.ylabel('y')
plt.title('Ajuste del modelo')
plt.legend()

# Gráfico de residuos
plt.subplot(1, 2, 2)
sns.residplot(x=model_sm.fittedvalues, y=model_sm.resid, lowess=True,
              line_kws={'color': 'red', 'lw': 2}, scatter_kws={'alpha': 0.7})
plt.xlabel('Valores ajustados')
plt.ylabel('Residuos')
plt.title('Residuos vs Valores ajustados')

plt.tight_layout()
plt.show()
```

```python
# Semilla para reproducibilidad
np.random.seed(42)

# Simulación de datos
n = 100
x = np.linspace(0, 10, n)
# Heterocedasticidad: varianza del error aumenta con x
errors = np.random.normal(0, 0.5 * x)
y = 2 + 0.8 * x + errors

# Convertir a DataFrame
df = pd.DataFrame({'x': x, 'y': y})

# Ajustar modelo
X = sm.add_constant(df['x'])  # agregar constante
model = sm.OLS(df['y'], X).fit()

# Resumen del modelo
print(model.summary())

# === Visualización ===

# Gráfico del ajuste y datos
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(df['x'], df['y'], label='Datos', alpha=0.7)
plt.plot(df['x'], model.fittedvalues, color='red', label='Ajuste')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Regresión lineal con heterocedasticidad')
plt.legend()

# Gráfico de residuos
plt.subplot(1, 2, 2)
sns.residplot(x=model.fittedvalues, y=model.resid, lowess=True,
              line_kws={'color': 'red'}, scatter_kws={'alpha': 0.7})
plt.xlabel('Valores ajustados')
plt.ylabel('Residuos')
plt.title('Residuos vs valores ajustados')

plt.tight_layout()
plt.show()
```
